# Analisis Data Pickle (baseline_bisindo.pkl)

Notebook ini bertujuan untuk mengekstrak dan menganalisis struktur data yang tersimpan di dalam file Pickle gabungan. Kita akan mengambil sampel 1 video dari masing-masing penutur (P01, P02, P03, P04, P05) dan menampilkan visualisasi frame pertamanya.

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Path ke file pickle utama (sesuaikan jika namanya berbeda)
pickle_file = 'data/pickle/baseline_bisindo.pkl'

print(f"Membaca data Pickle dari: {pickle_file}")
with open(pickle_file, 'rb') as f:
    data = pickle.load(f)

print(f"Total video dalam dataset gabungan: {len(data)}")
```

```python
# Mencari masing-masing 1 sampel dari tiap penutur (P01 - P05)
speakers = ["P01", "P02", "P03", "P04", "P05"]
selected_samples = {}

for video_id in data.keys():
    speaker_id = video_id.split('_')[0]
    if speaker_id in speakers and speaker_id not in selected_samples:
        selected_samples[speaker_id] = video_id
    if len(selected_samples) == 5:
        break

print("\n--- Sampel Terpilih (Pickle) ---")
for spk, vid in sorted(selected_samples.items()):
    keypoints = data[vid]['keypoints']
    print(f"Penutur: {spk} | Video ID: {vid} | Dimensi Array: {keypoints.shape} | Total Frame: {keypoints.shape[0]}")
```

```python
# Visualisasi sederhana frame ke-0 dari masing-masing sampel
if len(selected_samples) > 0:
    fig, axes = plt.subplots(1, len(selected_samples), figsize=(20, 4))
    if len(selected_samples) == 1:
        axes = [axes]
    
    for ax, (spk, vid) in zip(axes, sorted(selected_samples.items())):
        keypoints = data[vid]['keypoints']
        frame_0 = keypoints[0]
        
        # Ekstrak kordinat X dan Y
        x = frame_0[:, 0]
        y = frame_0[:, 1]
        
        # Sumbu Y di-invert agar sesuai dengan orientasi gambar aslinya
        ax.scatter(x, -y, s=10, c='blue', alpha=0.6)
        ax.set_title(f"{vid}\n({keypoints.shape[0]} frames)")
        ax.set_aspect('equal')
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("Belum ada sampel video yang ditemukan untuk divisualisasikan.")
```
