# Analisis Data Numpy (.npy)

Notebook ini bertujuan untuk melacak dan meload file `.npy` yang dihasilkan secara terpisah per video di folder `data/skeleton/`. Kita akan mengambil sampel 1 video dari masing-masing penutur (P01, P02, P03, P04, P05) dan menampilkan visualisasinya.

```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

npy_dir = Path('data/skeleton')
print(f"Mencari file .npy di dalam direktori: {npy_dir}")
```

```python
# Mencari masing-masing 1 sampel dari tiap penutur (P01 - P05)
speakers = ["P01", "P02", "P03", "P04", "P05"]
selected_files = {}

for npy_file in npy_dir.rglob("*.npy"):
    video_id = npy_file.stem
    speaker_id = video_id.split('_')[0]
    if speaker_id in speakers and speaker_id not in selected_files:
        selected_files[speaker_id] = npy_file
    if len(selected_files) == 5:
        break

print("\n--- Sampel Terpilih (Numpy) ---")
for spk, filepath in sorted(selected_files.items()):
    keypoints = np.load(filepath)
    video_id = filepath.stem
    # Mengambil path relatif agar tampilannya rapi
    rel_path = filepath.relative_to('data/skeleton')
    print(f"Penutur: {spk} | Video ID: {video_id} | Dimensi: {keypoints.shape} | Path: {rel_path}")
```

```python
# Visualisasi sederhana frame ke-0 dari masing-masing sampel Numpy
if len(selected_files) > 0:
    fig, axes = plt.subplots(1, len(selected_files), figsize=(20, 4))
    if len(selected_files) == 1:
        axes = [axes]
    
    for ax, (spk, filepath) in zip(axes, sorted(selected_files.items())):
        keypoints = np.load(filepath)
        video_id = filepath.stem
        
        frame_0 = keypoints[0]
        x = frame_0[:, 0]
        y = frame_0[:, 1]
        
        # Scatter plot kerangka (Y di-invert)
        ax.scatter(x, -y, s=10, c='green', alpha=0.6)
        ax.set_title(f"{video_id}\n({keypoints.shape[0]} frames)")
        ax.set_aspect('equal')
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("Belum ada file .npy yang ditemukan untuk penutur-penutur ini.")
```
